1. PANDAS (25 points)

```python
# 1 (2 marks). Load the netflix_titles.csv dataset as a Pandas dataframe:
import pandas as pd
netflix = pd.read_table('netflix_titles.csv', sep = ',')
netflix.head()
```

```python
# 2 (2 marks). Show the first 10 rows of the dataset
netflix.head(10)
```

```python
# 3 (1 mark). Show the last 5 rows
netflix.tail(5)
```

```python
# 4 (4 marks). Count the number of rows which have no missing value in any of the columns, and show the resulting data frame.
no_missing_values = netflix.isnull().sum()
print(no_missing_values)
print(netflix[['show_id', 'type', 'title', 'release_year', 'listed_in', 'description']].head())
```

```python
# 5 (3 marks). Drop everything that is not a 'Movie' and show the resulting data frame. Use any method for this.
print(netflix[netflix.type != 'Movie'].head(10))
```

```python
# 6 (7 marks). For every movie where the duration is missing, replace it with the average
  # duration of the movies in the dataset and return the resulting data frame
netflix['duration_minutes'] = netflix['duration'].str.extract('(\d+)').astype(float, '%.2f')
average_time = netflix[netflix.type == 'Movie']['duration_minutes'].mean()
print(average_time)
netflix['duration'] = netflix['duration'].fillna(average_time)
print(netflix.head(40))
```

```python
# 7 (3 marks). Are there any duplicates in the data? If yes, please remove them.
duplicates = netflix[netflix.duplicated()]
duplicates
```

```python
# 8 (2 marks). Rename all columns to start with a capital letter.
netflix.columns = netflix.columns.str.capitalize()
netflix
```

```python
# 9 (1 mark). Save the cleaned dataset as a new csv file.
netflix.to_csv('netflix_cleaned.csv')
```

2. MATPLOTLIB (25 points)

Use the netflix dataset to solve the exercises in this section.

```python
# 1 (10 marks). Create a visualisation (by using the plot that is most appropriate for this) which shows the frequency of movies by year of release.
import matplotlib.pyplot as plt

netflix_cleaned = pd.read_table('netflix_cleaned.csv', sep = ',')

year = netflix_cleaned['Release_year']
movies = netflix_cleaned[netflix_cleaned['Type'] == 'Movie']
frequency = movies.groupby('Release_year').size()
plt.plot(frequency.index, frequency.values, linewidth = 3)
plt.title('frequency of movies by year of release')
plt.ylabel('Frequency')
plt.xlabel('Year of Release')
plt.show()
```

```python
# 2 (15 marks). Create a pie chart which shows the 5 top most popular genres of the movis
  # in the dataset.

movies_by_genre = movies.groupby('Listed_in').size().sort_values(ascending = False)
movies_by_genre
top_genres = movies_by_genre.head(5)
genre_labels = top_genres.index
slices = top_genres.values

plt.pie(slices,
        labels=genre_labels,
        startangle=120,
        shadow=True,
        explode=(0, 0.2, 0, 0, 0),
        autopct='%1.1f%%')
plt.title('5 Most Popular Genres of Movie')
plt.show() 
```
